### Birthday
room = 365 people,
probability that 2 people have birthday in the same day

```python
k = 365
results = []

for n in range(1, k + 1):
    if n  == 1:
        results.append(1)
    else:
        x = results[-1] * (365 - n + 1) / 365
        results.append(x)


    
```

```python
import numpy as np
```

```python
res = 1 - np.array(results)
```

```python
fig, ax = plt.subplots()

ax.scatter(range(1, k + 1), res)

plt.show()
```

```python
# P(wszyscy maja roznego dnia ) = 365/365/ * 364/365 * ... * 365 - k +1 / 365
```
