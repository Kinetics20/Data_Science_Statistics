```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
num_pebbles = 1000
pebbles_x = np.random.rand(num_pebbles)
pebbles_y = np.random.rand(num_pebbles)
```

```python
event_A_center, event_A_radius = (0.4, 0.4), 0.3
event_B_center, event_B_radius = (0.6, 0.6), 0.3
```

```python
def is_inside_event(x, y, center, radius):
    return (x - center[0]) ** 2 + (y - center[1] ) ** 2 <= radius ** 2


pebbles_A = is_inside_event(pebbles_x, pebbles_y , event_A_center, event_A_radius)
pebbles_B = is_inside_event(pebbles_x, pebbles_y , event_B_center, event_B_radius)
```

```python
pebbles_A_and_B = pebbles_A & pebbles_B
```

```python
P_A = np.sum(pebbles_A) / num_pebbles
P_B = np.sum(pebbles_B) / num_pebbles
```

```python
P_A_and_P_B = np.sum(pebbles_A_and_B) / num_pebbles
P_A_and_P_B
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.scatter(pebbles_x, pebbles_y, color='grey', alpha=0.5, label='Pebbles')

circle_A = plt.Circle(event_A_center, event_A_radius, color='blue', alpha=0.3, label='Event A')
plt.gca().add_artist(circle_A)

circle_B = plt.Circle(event_B_center, event_B_radius, color='green', alpha=0.3, label='Event B')
plt.gca().add_artist(circle_B)


plt.legend()
plt.show()
```
